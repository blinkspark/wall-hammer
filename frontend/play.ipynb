```python
import yt_dlp
import yt_dlp.downloader
import json

with yt_dlp.YoutubeDL(
    {
        "dump_single_json": True,
        "skip_download": True,
        "writeinfojson": True,
        "outtmpl": "info",
    }
) as dl:
    res = dl.extract_info(
        "https://www.bilibili.com/video/BV1Pa411U7LD/?spm_id_from=333.1387.collection.video_card.click"
    )
    print(res)
    with open("info.json", "w", encoding="utf-8") as f:
        f.write(json.dumps(res, indent=2, ensure_ascii=False))
```
